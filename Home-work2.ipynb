```python
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt
# Use seaborn style defaults and set the default figure size
sns.set(rc={'figure.figsize':(15, 10)})
sns.set(font_scale=2)
sns.set_style("whitegrid")
sns.set_context("talk")
from pmdarima import auto_arima 
import pandas as pd 
import numpy as np
from statsmodels.tsa.stattools import acf, pacf
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf

from statsmodels.tsa.stattools import adfuller
from statsmodels.tsa.arima.model  import ARIMA
from statsmodels.tsa.seasonal import seasonal_decompose
```

```python
df = pd.read_csv('../datatest/monthly-car-sales.csv')
```

```python
df.head()
```

```python
df['Month'] = pd.to_datetime(df['Month'])
df.set_index('Month',inplace=True)
df.index = pd.DatetimeIndex(df.index.values, freq=df.index.inferred_freq)
df['Sales'].plot()
```

```python
result=seasonal_decompose(df['Sales'], model='multiplicative', period=12)
result.trend.plot()
```

```python
result = adfuller(df['Sales'])
print('p-value: %f' % result[1])
```

```python
df['diff'] = df['Sales'].diff()
df.dropna(inplace=True)

result = adfuller(df['diff'])
print('p-value: %f' % result[1])
```

```python
df['diff'].plot()
```

```python
# split data into train and test

# what is the size of 80% ? 
size = int(len(df)*0.8)

# use 80% for train and 20% for test

train, test = df[0:size], df[size:]
print(f'Length of train is {len(train)} and lenght of test is {len(test)}')
```

```python
model = ARIMA(train['Sales'], order=(2,1,2))
model_fit = model.fit(cov_type = 'approx')
preds = model_fit.forecast(len(test))
```

```python
preds.plot()
test['Sales'].plot()
plt.show()
```

 <code>auto_arima</code>    ,     (  )     .      ,   <code> r2_score </code>  <code> mean_absolute_error </code>.

```python
from sklearn.metrics import r2_score, mean_absolute_error
```

```python
model = auto_arima(train['Sales'],
                   start_p=1,
                   start_q=1,
                   max_p=8,
                   max_q=8,
                   start_P=0,
                   start_Q=0,
                   max_P=8,max_Q=8,
                   m=12,seasonal=True,
                   stepwise=False,
                   suppress_warnings=True,D=1,d=1,error_action='warn',random_state=20,n_fits=30)
```

```python
model.summary()
```

```python
model_fit = model.fit(train['Sales'])
preds = model_fit.predict(len(test))
```

```python
predictions = pd.DataFrame(preds)
```

```python
predictions.index = test.index
```

```python
predictions.plot()
test['Sales'].plot()
plt.show()
```

```python
r2_score(test['Sales'], predictions)
```

```python
mean_absolute_error(test['Sales'], predictions)
```
